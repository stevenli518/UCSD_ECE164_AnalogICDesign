```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Vo_sat =  [0, 1, 2, 3, 4, 4.5, 4.8, 4.9]
Id = []
Vsd = []
for vo in Vo_sat:
    Vsd.append((5.5-vo))
    id = 0.5* 40e-6*0.25*(0.6**2)*(1+0.1*(5.5-vo))
    Id.append(id*1000000)
    print(vo, (5.5-vo), id*1000000)

Vo_tri = [5.1, 5.2, 5.35, 5.5]
for vo in Vo_tri:
    Vsd.append((5.5-vo))
    id = 40e-6 * 0.25 * (0.6 * (5.5-vo) - 0.5*((5.5-vo)**2))
    Id.append(id*1000000)
    print(vo, (5.5-vo), id*1000000)
```

```python
plt.plot(Vo_sat+ Vo_tri, Id)
plt.xticks(np.arange(0, 6, 0.5))  # Set x-axis ticks at intervals of 1
plt.yticks(np.arange(0, 3, 0.2))
plt.title("I_D(uA) vs V_O(V)")
plt.xlabel("V_O(V)")
plt.ylabel("I_D(uA)")
plt.show()
```
